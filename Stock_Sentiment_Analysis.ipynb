## Stock Sentiment Analysis 

**Stock Sentiment Analysis using headlines**

*   The dataset is from Kaggle (https://www.kaggle.com/aaron7sun/stocknews)
*   Labels are based on Dow Jones Induestrial Average Stock Index 
*   Label 0 : Stock Price increased
*   Label 1 : Stock Price stayed the same or decreased
*   I used few Classification methods and calucated their accuracy scores.






```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Data.csv', encoding = "ISO-8859-1")
```

```python
#df.head()
df.tail()
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
test.shape
test.head()
```

```python
# Removing punctuations
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)

```

```python
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implementing BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
# implementing RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
# implementing DecisionTree Classifier
from sklearn.tree import DecisionTreeClassifier
StockTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
StockTree
```

```python
StockTree.fit(traindataset,train['Label'])
```

```python
# implementing SVM
from sklearn import svm
SVM_clf = svm.SVC(kernel='rbf')
SVM_clf.fit(traindataset,train['Label']) 
```

```python
# implementing LogisticRegression
from sklearn.linear_model import LogisticRegression
Log_reg = LogisticRegression(C=0.01, solver='liblinear').fit(traindataset,train['Label'])
```

```python
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
# Prediction of RandomForest Classifier
RandomClassifier_pred = randomclassifier.predict(test_dataset)
```

```python
# Prediction of DecisionTree Classifier
StockTree_pred = StockTree.predict(test_dataset)
```

```python
# Prediction of SVM
SVM_pred = SVM_clf.predict(test_dataset)
```

```python
# Prediction of LogisticRegression
Log_reg_pred = Log_reg.predict(test_dataset)
```

```python
# Prediction of KNN
KNN_pred = KNN_clf.predict(test_dataset)
```

```python
## Importing library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
# Scores of RandomForest Classifier
print("Scores of RandomForest Classifier")
matrix=confusion_matrix(test['Label'],RandomClassifier_pred)
print(matrix)
score=accuracy_score(test['Label'],RandomClassifier_pred)
print(score)
report=classification_report(test['Label'],RandomClassifier_pred)
print(report)
```

```python
# Scores of DecisionTree Classifier
print("Scores of Decision Classifier")
matrix=confusion_matrix(test['Label'],StockTree_pred)
print(matrix)
score=accuracy_score(test['Label'],StockTree_pred)
print(score)
report=classification_report(test['Label'],StockTree_pred)
print(report)
```

```python
# Scores of SVM
print("Scores of SVM")
matrix=confusion_matrix(test['Label'],SVM_pred)
print(matrix)
score=accuracy_score(test['Label'],SVM_pred)
print(score)
report=classification_report(test['Label'],SVM_pred)
print(report)
```

```python
# Scores of Logistic Regression
print("Scores of Logistic Regression")
matrix=confusion_matrix(test['Label'],Log_reg_pred)
print(matrix)
score=accuracy_score(test['Label'],Log_reg_pred)
print(score)
report=classification_report(test['Label'],Log_reg_pred)
print(report)
```

```python
# Scores of KNN
print("Scores of KNN")
matrix=confusion_matrix(test['Label'],KNN_pred)
print(matrix)
score=accuracy_score(test['Label'],KNN_pred)
print(score)
report=classification_report(test['Label'],KNN_pred)
print(report)
```
